# **Legenda**

Age = idade (anos)

Sex = sexo (0=M; 1=F)

Chest Pain Type = tipo de dor no peito (0=TA:angina tpica; 1=ATA:angina atpica; 2=NAP:dor no anginosa; 3=ASY:assintomtico)

Resting BP = presso sangunea em repouso (mmHg)

Cholesterol = colesterol srico (mg/dl)

Fasting BS = acar no sangue em jejum (mg/dl) 0: Fasting BS < 120 mg/dl (no diabtico) 1: Fasting BS >= 120 mg/dl, (diabtico)

Resting ECG = eletrocardiograma em repouso (0=Normal; 1=ST:Anormalidade da onda ST-T; 2=LVH: Hipertrofia ventricular esquerda)

Max HR = frequncia cardaca mxima

Exercise Angina = Angina induzida por exerccio (0=No; 1=Sim)

Old Peak = Depresso de ST induzida por exerccio em relao ao repouso

ST_Slope = Inclinao do segmento ST (0=UP; 1=Flat; 2=Down)

Heart Disease = Doena cardaca (0=No possui doena cardaca; 1=Possui doena cardaca)

# **OneHotEncoder: Criao de variveis Dummy (fictcia)**

Variveis criadas para evitar a existncia de outras variveis altamente correlacionadas entre si.

A = 0, B = 1, C = 2, D = 3


A   B   C   D

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn. compose import ColumnTransformer
```

```python
ColumnTransformer?
```

Parmetros ColumnTransformer
- name: nome dado a transformao.
- transformer: tipo de estimador (OneHotEncoder).
- columns: colunas que sero transformadas.
- remainder: o que acontecer com o restante das colunas no relacionadas:
1) drop = exclui as outras colunas.
2) passthrough = mantm as outras colunas. drop  default.  
- sparse_threshold: parmetro de classificao de matrizes esparsas. default  0.3
- n_jobs: nmero de trabalhos a serem executados em paralelo. default  nenhum
- transformer_weights: definio de pesos aos transformadores.
-verbose: default  False. se for True a execuo  apresentada na tela.

```python
previsores3 = ColumnTransformer(transformers=[('OneHot', OneHotEncoder(), [1,2,6,8,10])],
                                remainder='passthrough').fit_transform(previsores2)
# estabelecendo o nome da transformao, tipo de transformao, colunas que sero transformadas, colunas que sero mantidas ou excludas.
```

```python
previsores3               # apresentando a padronizao de todas as linhas e colunas que precisavam de transformao nmerica
```

Parmetros train_test_split:   
- arrays: nomes dos atributos previsores e alvo.   
- test_size: tamanho em porcentagem dos dados de teste. default  none.   
- train_size: tamanho em porcentagem dos dados de treinamento.default  none.  
- random_state: nomeao de um estado aleatrio.   
- shuffle: embaralhamento dos dados aleatrios. Associado com o random_state ocorre o mesmo embaralhamento sempre. Default  True.  
- stratify: Possibilidade de dividir os dados de forma estratificada. Default  None (nesse caso  mantido a proporo, isto , se tem 30% de zeros e 70% de 1 no dataframe, na separao em treinamento e teste se manter essa proporo).
